```python
import pandas as pd
import pymongo

class CSVToMongoDB:
    def __init__(self, mongodb_uri: str, db_name: str, collection_name: str):
        self.mongodb_uri = mongodb_uri
        self.db_name = db_name
        self.collection_name = collection_name
        self.client = pymongo.MongoClient(self.mongodb_uri)
        self.db = self.client[self.db_name]
        self.collection = self.db[self.collection_name]

    def ingest_csv(self, csv_file_path: str):
        """
        Ingest a CSV file into the MongoDB collection.

        Args:
        csv_file_path (str): The path to the CSV file.
        """
        try:
            # Read the CSV file into a DataFrame
            df = pd.read_csv(csv_file_path)

            # Convert DataFrame to a list of dictionaries
            data = df.to_dict(orient='records')

            # Insert data into MongoDB collection
            self.collection.insert_many(data)
            print(f"Successfully ingested {len(data)} records into MongoDB.")
        except Exception as e:
            print(f"Error ingesting CSV to MongoDB: {e}")
```

```python
from sentence_transformers import SentenceTransformer

# https://huggingface.co/thenlper/gte-large
embedding_model = SentenceTransformer("NeuML/pubmedbert-base-embeddings")


# ngoc
# vvbnESrLE2gcbkmW
# mongodb+srv://ngoc:vvbnESrLE2gcbkmW@cluster0.jps061j.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0

def get_embedding(text: str) -> list[float]:
    if not text.strip():
        print("Attempted to get embedding for empty text.")
        return []

    embedding = embedding_model.encode(text)

    return embedding.tolist()
```

```python
def embed_rows_and_add_column(csv_file, output_file):
    # Step 1: Load CSV data
    df = pd.read_csv(csv_file)
    
    # Step 3: Create embeddings for each row
    embeddings = []
    for text in df['overview']:
        # Concatenate all column values into a single string
        #row_text = " ".join(row.astype(str))
        
        # Generate embedding for the concatenated row content
        embedding = get_embedding(text)
        
        # Append the embedding as a list (for easier storage in CSV later)
        embeddings.append(embedding)
    
    # Step 4: Add the embeddings as a new column to the DataFrame
    df['embedding'] = embeddings
    
    # Step 5: Save the modified DataFrame to a new CSV file
    df.to_csv(output_file, index=False)
    
    print(f"Embeddings added and saved to {output_file}!")
```

```python
# Example usage:
csv_file = "data\companies.csv"
output_file = "data\output_with_embeddings.csv"

dataset_df=pd.read_csv(csv_file)
dataset_df["embedding"] = dataset_df["overview"].apply(get_embedding)

embed_rows_and_add_column(csv_file, output_file)
```
